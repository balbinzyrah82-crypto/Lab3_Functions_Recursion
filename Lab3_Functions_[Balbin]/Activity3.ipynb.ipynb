```python
# [CELL 1] STUDENT CONFIGURATION

LAST_NAME = "Balbin"            
STUDENT_ID = "TUPM-24-0561"     

SEED_DIGIT = int(STUDENT_ID[-1])
ID_SUM = sum(int(d) for d in STUDENT_ID if d.isdigit())
NAME_LENGTH = len(LAST_NAME)

print("Student:", LAST_NAME)
print("Seed Digit:", SEED_DIGIT)
print("Digit Sum:", ID_SUM)
print("Surname Length:", NAME_LENGTH)
```

```python
# [CELL 2] FUNCTIONAL ENCAPSULATION

def greet():
    print("=" * 40)
    print(f"System Initialized for: {LAST_NAME}")
    print("=" * 40)

def identity_code():
    code = SEED_DIGIT * NAME_LENGTH
    return code


greet()
print(f"Your Identity Code is: {identity_code()}")

```

```python
# [CELL 3] LEGB VALIDATION

global_value = ID_SUM

def scope_test():
    global_value = SEED_DIGIT
    print("Inside function:", global_value)

scope_test()
print("Outside function:", global_value)
```

```python
# [CELL 4] PARAMETRIC PROCESSING

def user_summary(title, *scores, **info):
    print(f"=== {title} ===")
    
    total = sum(scores)
    average = total / len(scores)
    
    print("Scores:", scores)
    print("Total:", total)
    print("Average:", round(average, 2))
    
    for k, v in info.items():
        print(f"{k}: {v}")
        
    return average


avg = user_summary(
    f"{LAST_NAME} Academic Report",
    SEED_DIGIT, 10, 85, 90,
    Subject="ECE Fundamentals",
    Status="Active"
)
```

```python
# [CELL 5] RETURN MECHANISM - Task 8
def compute_area(radius):
    return 3.1416 * radius**2

area = compute_area(SEED_DIGIT + 2)
print("\n--- Task 8: Return Mechanism ---")
print("Computed Area:", round(area, 2))


```

```python
# [CELL 6] SAFE CALCULATOR - Task 9
def safe_calculator(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        result = "UNDEFINED"
    return result

print("\n--- Task 9: Fault-Tolerant Arithmetic ---")
print("Calculation Result:", safe_calculator(ID_SUM, SEED_DIGIT))


```

```python
# [CELL 7] RECURSION - Task 10
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print("\n--- Task 10: Recursion ---")
print(f"Factorial of {SEED_DIGIT}:", factorial(SEED_DIGIT))
```

```python
# [CELL 8] STRING RECURSION - Task 11
def reverse(text):
    if len(text) <= 1:
        return text
    return text[-1] + reverse(text[:-1])

print("\n--- Task 11: String Recursion ---")
print("Reversed Surname:", reverse(LAST_NAME.upper()))
```

```python
# [CELL 9] DECORATOR - Task 12
def logger(func):
    def wrapper(*args, **kwargs):
        print("Executing:", func.__name__)
        return func(*args, **kwargs)
    return wrapper

@logger
def multiply(x, y):
    return x * y

print("\n--- Task 12: Python Decorators ---")
print("Multiplication:", multiply(SEED_DIGIT, NAME_LENGTH))
```

```python
# [CELL 10] LAMBDA - Task 13
data = list(range(1, SEED_DIGIT + 10))
squared = list(map(lambda x: x**2, data))
print("\n--- Task 13: Lambda Functions ---")
print("Squared Data:", squared)

```

```python
# [CELL 11] GENERATOR - Task 14
def even_squares(limit):
    for i in range(limit):
        if i % 2 == 0:
            yield i**2

print("\n--- Task 14: Generator Implementation ---")
for value in even_squares(NAME_LENGTH + SEED_DIGIT):
    print(value) 
```
